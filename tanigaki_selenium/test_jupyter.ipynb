```python
import time
from selenium import webdriver
import chromedriver_binary
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait

import openpyxl
```

```python
driver = webdriver.Chrome()
driver.get("http://localhost:5500")

iframe_css = ".iframe"
elem_css = "p"

wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, iframe_css)))
iframe = driver.find_element(By.CSS_SELECTOR, iframe_css)
```

```python
# iframeの中に入る
driver.switch_to.frame(iframe)
# iframeの中の要素のテキストを取得
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, elem_css)))
element = driver.find_element(By.CSS_SELECTOR, elem_css)
txt = element.text
# iframeの中から出る
driver.switch_to.default_content()

print(txt)
```
